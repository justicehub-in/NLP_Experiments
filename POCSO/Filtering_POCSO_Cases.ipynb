# POCSO FILTER

There are 1764 POCSO Judgments available To filter the POCSO cases upon which NER will be implemented.

Filter criteria:
1. State (state_name): Assam
2. Registered year (reg_year_date): 2017-2019
3. Disposal code (disp_nature): All Disposed cases. (All codes except 0)

```python
#Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pocso_df = pd.read_csv('pocso_analysis_datasets_1221/all_pocso_cases.csv')
pocso_df.head()
```

Data source: [JusticeHub](https://justicehub.in/dataset/data4justice-unpacking-judicial-data-to-track-implementation-of-the-pocso-act-in-assam-delhi-haryana/resource/f667d254-0f39-4bd3-8f6a-e3043d0991f1)

```python
# Filter 1
## POCSO cases in Assam
## POCSO cases registered in the years 2017, 2018, 2019
## POCSO cases that are disposed (Judgement Delivered)
pocso_df_filter1 = pocso_df[(pocso_df.state_name =='Assam')&
                            (pocso_df.reg_year_date.isin([2017, 2018, 2019]))&
                            (pocso_df.disp_nature!=0)]
pocso_df_filter1.head()
```

```python
pocso_df_filter1.shape
```

```python
pocso_df_filter1.to_csv('Results/POCSO_Filter1.csv',index=False)
```

There are 1764 cases based on the first filter. More filter criterias would be based on:
1. Case Type (updated_type_name)
2. Districts (district_name)

```python
# EDA - District wise number of judgements delivered
pocso_df_filter1.groupby('district_name', as_index=False).size().sort_values('size').set_index('district_name').plot(kind='barh', figsize=(5,7), legend=None)
plt.ylabel('District')
plt.xlabel('Number of cases')
plt.show()
```

```python
# EDA - CaseType wise number of judgements delivered
pocso_df_filter1.groupby('updated_type_name', as_index=False).size().sort_values('size').set_index('updated_type_name').plot(kind='barh', figsize=(4,5), legend=None)
plt.ylabel('Case Type')
plt.xlabel('Number of cases')
plt.show()
```

```python
# District - CaseType heatmap
pocso_df_filter1_heatmap = pocso_df_filter1.pivot_table(values ='cino', index='district_name', columns='updated_type_name', aggfunc='count')
pocso_df_filter1_heatmap = pocso_df_filter1_heatmap.fillna(0)
pocso_df_filter1_heatmap = pocso_df_filter1_heatmap.astype(np.int64)

pocso_df_filter1_heatmap.style.background_gradient(cmap ='Purples').set_properties(**{'font-size': '15px'})
```

A combination of case types and districts should be decided to further filter the POCSO cases

Second set of filters are applied as follows:
1. Select 10 `Special (POCSO) Case` cases from each of Barpeta, Sivasagar and Sonitpur
2. Select 10 `POCSO Act` cases from each of Sonitpur and Nagaon

But not every disposed case has a judgement available. So we take more cases so that we have a sample of 50 judgement copies.

```python
pocso_df_filter2_specialpocso = pocso_df_filter1[(pocso_df_filter1.updated_type_name=='Special (POCSO) Case')&
                                                 (pocso_df_filter1.district_name.isin(['Barpeta','Sivasagar','Sonitpur']))].sample(75)
pocso_df_filter2_pocsoact = pocso_df_filter1[(pocso_df_filter1.updated_type_name=='POCSO Act')&
                                             (pocso_df_filter1.district_name.isin(['Sonitpur', 'Nagaon']))].sample(50)
```

```python
pocso_df_filter2 = pd.concat([pocso_df_filter2_specialpocso,pocso_df_filter2_pocsoact]).reset_index(drop=True)
pocso_df_filter2.tail()
```

Out of these 125 disposed cases, only 51 have judgement PDFs available on eCourts.

```python
pocso_df_filter2.to_csv('Results/POCSO_Filter2.csv',index=False)
```
